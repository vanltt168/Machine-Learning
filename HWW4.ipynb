```python
covid19 = pd.read_csv('/Users/macbook/Downloads/covid19_cases.csv')
covid19.head()
```

```python
#b. Create a date column using the data in the dateRep column and the pd.to_datetime() function.
covid19['date'] = pd.to_datetime(covid19['dateRep'])
covid19.head()
```

```python
#c. Set the date column as the index and sort the index
covid19.set_index('dateRep', inplace=True) 
covid19.head()
```

```python
covid19.sort_index()
covid19.head()
```

```python
#d
covid19.replace('United_States_of_America', 'USA')
covid19.replace('United_Kingdom', 'UK')
```
